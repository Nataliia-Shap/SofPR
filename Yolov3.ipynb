```python
! pip install imageai
from imageai.Detection import VideoObjectDetection
import os

execution_path = os.getcwd()

detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join('/content/drive/My Drive/Colab Notebooks', 'yolo.h5'))
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path=os.path.join("/content/drive/My Drive/Colab Notebooks/", "traffic.mp4"), 
                            output_file_path=os.path.join("/content/drive/My Drive/Colab Notebooks/", "traffic_detected")
                            , frames_per_second=20, log_progress=True)
print(video_path)
```
